```python
import plotly.plotly as py

import urllib, json

url = 'https://raw.githubusercontent.com/plotly/plotly.js/master/test/image/mocks/sankey_energy.json'
response = urllib.request.urlopen(url)
data = json.loads(response.read())

node_label = data['data'][0]['node']['label']
color = data['data'][0]['node']['color']
source =  data['data'][0]['link']['source']
target =  data['data'][0]['link']['target']
value =  data['data'][0]['link']['value']
label =  data['data'][0]['link']['label']

data_trace = dict(
    type='sankey',
    width = 1118,
    height = 772,
    domain = dict(
      x =  [0,1],
      y =  [0,1]
    ),
    orientation = "h",
    valueformat = ".0f",
    valuesuffix = "TWh",
    node = dict(
      pad = 15,
      thickness = 15,
      line = dict(
        color = "black",
        width = 0.5
      ),
      label =  data['data'][0]['node']['label'], # Name of a source or target node.
      color =  "blue" # Color to associate with it.
    ),
    link = dict( # Controls edges within the diagram
      source =  data['data'][0]['link']['source'], # Node Index  node_label[source]
      target =  data['data'][0]['link']['target'], # Node Index. node_label[target] 
      value =  value, # Weight of the edge (controls the width)
      label =  data['data'][0]['link']['label'] # Name of the edge when hovered over.
  ))

layout =  dict(
    title = "Energy forecas",
    font = dict(
      size = 10
    )
)

fig = dict(data=[data_trace], layout=layout)
# py.iplot(fig, validate=False)
plotly.offline.iplot(fig, validate=False)
```
